```python
import matplotlib.pyplot as plt
```

```python
def draw_bmh_style():
    plt.style.use("bmh")
    plt.plot([1, 3, 9, 5, 2, 1, 1], marker='o')
    plt.plot([4, 5, 5, 7, 9, 8, 6], marker='v')
    plt.plot([2, 3, 4, 3, 4, 5, 3], marker='s')
    plt.show()
    
def draw_fivethirtyeight():
    plt.style.use("fivethirtyeight")
    plt.plot([1, 3, 9, 5, 2, 1, 1], marker='o')
    plt.plot([4, 5, 5, 7, 9, 8, 6], marker='v')
    plt.plot([2, 3, 4, 3, 4, 5, 3], marker='s')
    plt.show()

def draw_ggplot():
    plt.style.use("ggplot")
    plt.plot([1, 3, 9, 5, 2, 1, 1], marker='o')
    plt.plot([4, 5, 5, 7, 9, 8, 6], marker='v')
    plt.plot([2, 3, 4, 3, 4, 5, 3], marker='s')
    plt.show()

def draw_classic():
    plt.style.use("classic")
    plt.plot([1, 3, 9, 5, 2, 1, 1], marker='o')
    plt.plot([4, 5, 5, 7, 9, 8, 6], marker='v')
    plt.plot([2, 3, 4, 3, 4, 5, 3], marker='s')
    plt.show()

def draw_default_style():
    #plt.style.use("classic")
    plt.plot([1, 3, 9, 5, 2, 1, 1], marker='o')
    plt.plot([4, 5, 5, 7, 9, 8, 6], marker='v')
    plt.plot([2, 3, 4, 3, 4, 5, 3], marker='s')
    plt.show()    
```

```python
draw_bmh_style()
```

```python
draw_fivethirtyeight()
```

```python
draw_ggplot()
```

```python
draw_classic()
```

```python
draw_default_style()
```
